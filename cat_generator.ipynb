```python
import numpy as np 
from tensorflow.keras.layers import Dense , Flatten , Reshape
from tensorflow.keras.models import Model , Sequential 
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img , img_to_array
```

```python
net = Sequential()
net.add(Dense(50 , activation='sigmoid' , input_shape=(50,)))
net.add(Dense(50 , activation='relu'))
net.add(Dense(120000 , activation='relu'))
net.add(Reshape((200 , 200,3) ))
net.compile(optimizer='adam' , loss='mean_squared_error')
```

```python
s = np.random.randn(1 , 50)
real_img = load_img('./pic0.jpg'  , target_size=(200,200))
image = np.array([img_to_array(real_img)])
print(image.shape)
```

```python
h = net.fit(s , image , epochs=500)
```

```python
s = np.random.randn(1 , 50) +10
a = net.predict(s)
pic = a[0 , : ,: ,:] *1
pic = pic.astype(np.uint8)
_ , ax = plt.subplots(1 , 2)
ax[0].imshow(pic)
ax[0].set_title('genetated from noise')
ax[1].imshow(image[0,:,:,:].astype(np.uint8))
ax[1].set_title('real image')
plt.show()
net.save('cat_generator.h5')
```
